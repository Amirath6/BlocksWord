```python
variables30 = df.values[20:30]
```

```python
variables40 = df.values[30:40]
```

```python
variables50 = df.values[40:]
```

```python
moyenne30 = np.mean(variables30, axis=0, dtype=np.int64)
```

```python
moyenne40 = np.mean(variables40, axis=0, dtype=np.int64)
```

```python
moyenne50 = np.mean(variables50, axis=0, dtype=np.int64)
```

```python
result = np.stack((moyenne10, moyenne20, moyenne30, moyenne40, moyenne50), axis=0)
```

```python
print(result.shape)
print(result)
```

```python
plt.plot(result[:,0], np.log(result[:,1]), label="bfs")
plt.plot(result[:,0], np.log(result[:,3]), label="dfs")
plt.plot(result[:,0], np.log(result[:,5]), label="dijk")
plt.plot(result[:,0], np.log(result[:,7]), label="A*")

plt.legend()
plt.show()
```

```python
plt.plot(result[:,0], result[:,2], label="bfs")
plt.plot(result[:,0], result[:,4], label="dfs")
plt.plot(result[:,0], result[:,6], label="dijk")
plt.plot(result[:,0], result[:,8], label="A*")
plt.legend()
plt.show()
```
